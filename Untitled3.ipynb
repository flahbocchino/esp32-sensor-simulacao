```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = """
Temperatura: 24.00 °C  |  Umidade: 40.00 %
Temperatura: 24.00 °C  |  Umidade: 40.00 %
Temperatura: 24.00 °C  |  Umidade: 40.00 %
Temperatura: 24.00 °C  |  Umidade: 40.00 %
Temperatura: 24.00 °C  |  Umidade: 40.00 %
"""  # Cole aqui mais linhas, se quiser
```

```python
linhas = dados.strip().split('\n')
temperaturas = []
umidades = []

for linha in linhas:
    if "Temperatura" in linha and "Umidade" in linha:
        partes = linha.split('|')
        temp = float(partes[0].split(':')[1].replace('°C', '').strip())
        umi = float(partes[1].split(':')[1].replace('%', '').strip())
        temperaturas.append(temp)
        umidades.append(umi)

df = pd.DataFrame({
    'Tempo (s)': list(range(len(temperaturas))),
    'Temperatura (°C)': temperaturas,
    'Umidade (%)': umidades
})

df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Tempo (s)'], df['Temperatura (°C)'], label='Temperatura (°C)')
plt.plot(df['Tempo (s)'], df['Umidade (%)'], label='Umidade (%)')
plt.title('Dados do sensor DHT22 simulados no Wokwi')
plt.xlabel('Tempo (s)')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```
